```python
import numpy as np 
import pandas as pd 
```

```python
july_weather = pd.read_csv('/Users/quynhanh/Documents/SEMESTER/SEM5/Big Data/nyc_yellow_taxi_eta/data/weather_data/New York City,USA 2024-07-01 to 2024-07-31.csv')
august_weather = pd.read_csv('/Users/quynhanh/Documents/SEMESTER/SEM5/Big Data/nyc_yellow_taxi_eta/data/weather_data/New York City,USA 2024-08-01 to 2024-08-31.csv')
september_weather = pd.read_csv('/Users/quynhanh/Documents/SEMESTER/SEM5/Big Data/nyc_yellow_taxi_eta/data/weather_data/New York City,USA 2024-09-01 to 2024-09-30.csv')

july_weather.head()
```

```python
merged_weather = pd.concat([july_weather, august_weather, september_weather], ignore_index=True)
merged_weather.head()
```

```python
merged_weather['datetime'] = pd.to_datetime(merged_weather['datetime'])

merged_weather['date'] = merged_weather['datetime'].dt.date
merged_weather['time'] = merged_weather['datetime'].dt.time

merged_weather['day'] = merged_weather['datetime'].dt.day
merged_weather['month'] = merged_weather['datetime'].dt.month
merged_weather['hour'] = merged_weather['datetime'].dt.hour

merged_weather.head()
```

```python
merged_weather.drop(['feelslike', 'dew', 'precipprob', 'preciptype', 'snow', 'conditions', 'icon', 'stations', 'uvindex', 'sealevelpressure', 'solarenergy', 'solarradiation', 'snowdepth'], axis=1, inplace=True)
merged_weather.head()
```

```python
merged_weather.to_csv('merged_weather.csv', index=False)
```
